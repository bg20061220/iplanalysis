```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

delivery = pd.read_csv('/kaggle/input/ipl/deliveries.csv')
df = pd.read_csv('/kaggle/input/ipl/matches.csv')

matches_rows = df.shape[0]
# print(f"Matches Rows:  {matches_rows}")
# Question 1 Role of Toss in Winning Matches  Ans : 51 % 
#counter = 0 ;
#for index , row in matches.iterrows() :
 #  if row['toss_winner'] == row['winner'] :
  #     counter = counter + 1  
#print(f"Teams who won the toss that went onto win the match were ; {counter/matches_rows}") 

# Question 2 Is the percentage same when team takes batting first and bowling first 
# Bowling First = 55.37 %  Batting First = 45.42% 
# print(df.head(60))
#won_toss_field = matches[matches['toss_decision'] == 'bat'] 
#won_toss_field_and_match = won_toss_field[won_toss_field['toss_winner'] == won_toss_field['winner']]
#total_toss_field = len(won_toss_field)
#total_won_field_and_match = len(won_toss_field_and_match)
#percentage = (total_won_field_and_match / total_toss_field) * 100 
#print(f"Teams who went onto win toss, chose too field are {percentage}")

#Question 3 Finding Out if teams win more chasing or defending 
# Ans Win Chasing = 54.8742% and Win Defending = 45.12% 
#win_chasing = 0 
#win_defending = 0
#for index , row in matches.iterrows():
 #   if row['win_by_runs'] == 0 :
 #       win_chasing += 1 
 #   else :
 #       win_defending += 1 
#win_chasing_percentage = (win_chasing / matches_rows) * 100 
#win_defending_percentage = (win_defending / matches_rows) * 100 
#print(f"Win Chasing : {win_chasing_percentage} , Win Defending : {win_defending_percentage}")

# Q4 As the competition goes does the chasing teams starts to lose
# Ans : Yes as competition goes chances of defending team chances to win increases,  Equation = 0.30x + 36.51 because as tournament goes on pitch get worse
# and it is difficult to chase.
# season_df = df[df['id'] <= 59]
# season_df['defending_won'] = season_df['win_by_runs'] > 0
# season_df['match_interval'] = (season_df['id'] // 10) * 10
# defending_stats = season_df.groupby('match_interval')['defending_won'].mean().reset_index()
# defending_stats['defending_win_percentage'] = defending_stats['defending_won'] * 100
print(df.columns)
# sns.lmplot(x='match_interval', y='defending_win_percentage', data=defending_stats, aspect=1.5, ci=None)
# plt.title('Trend of Defending Teams Winning (First 59 Matches)')
# plt.xlabel('Match Interval')
# plt.ylabel('Defending Win Percentage')
# plt.show()
# X = defending_stats['match_interval'].values.reshape(-1, 1)  # Match intervals (independent variable)
# y = defending_stats['defending_win_percentage'].values      # Win percentages (dependent variable)
# model = LinearRegression()
# model.fit(X, y)
# slope = model.coef_[0]
# intercept = model.intercept_
# print(f"Equation of the line: y = {slope:.2f}x + {intercept:.2f}")

#Q5) Finding out how teams perform at their homeGround
matches_2017 = df[df['season'] == 2017] 
unique_venues_2017 = matches_2017['venue'].unique()
print(unique_venues_2017)
unique_teams = set(matches_2017['team1'].unique()).union(set(matches_2017['team2'].unique()))
print(unique_teams)
team_home_ground = {
   'Kolkata Knight Riders': ['Eden Gardens'],
    'Mumbai Indians': ['Wankhede Stadium'],
    'Royal Challengers Bangalore': ['M Chinnaswamy Stadium'],
    'Rising Pune Supergiant': ['Maharashtra Cricket Association Stadium'],
    'Delhi Daredevils': ['Feroz Shah Kotla'],
    'Gujarat Lions': ['Saurashtra Cricket Association Stadium' , 'Green Park'],
    'Sunrisers Hyderabad': ['Rajiv Gandhi International Stadium, Uppal'],
    'Kings XI Punjab': ['Punjab Cricket Association IS Bindra Stadium, Mohali' , 'Holkar Cricket Stadium']
}
def is_home_game(team, venue, team_home_ground):
    return venue in team_home_ground.get(team, [])
df['is_home_game_team1'] = df.apply(lambda row: is_home_game(row['team1'], row['venue'], team_home_ground), axis=1)
df['is_home_game_team2'] = df.apply(lambda row: is_home_game(row['team2'], row['venue'], team_home_ground), axis=1)
df['is_home_team_win'] = ((df['is_home_game_team1'] & (df['winner'] == df['team1'])) | 
                          (df['is_home_game_team2'] & (df['winner'] == df['team2'])))
home_performance = df.groupby('winner')['is_home_team_win'].mean().reset_index()
home_performance.columns = ['team', 'home_win_percentage']
home_performance['home_win_percentage'] *= 100

print(home_performance)
# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```
